```python
def get_kolesa(page):
    cars = []
    links = []
    url = "https://kolesa.kz/cars/bmw/?year[from]=2016&year[to]=2018&page="+str(page)
    page = requests.get(url)
    soup = BeautifulSoup(page.content,"html.parser")
    items = soup.find_all("a", class_="ddl_product_link")
    for item in items:
        images = item.find_all("img")
        if len(images)>0:
            links.append(item.attrs["href"])
            first_img = images[0]
            cars.append(str(first_img.attrs["alt"]))
    return cars, links
```

```python
page = 1
all_cars = []
all_links = []
while True:
    print("Parcing page",page)
    cars, links = get_kolesa(page)
    if len(cars)==0:
        break
    all_cars += cars
    all_links += links
    page +=1
    
```

```python
price = [None]*len(all_cars)
for i in range(len(all_cars)):
    s = all_cars[i]
    result = re.search("за (.*) тг.", s)
    price[i] = int(result.group(1).replace(" ",""))
```

```python
def the_price(link):
    page = requests.get("https://kolesa.kz"+link)
    content = page.content.decode("utf-8").lower()
    soup = BeautifulSoup(page.content, "html.parser")
    return soup.find_all("div", class_="offer__price")[0].contents[0].replace(" ","")
```

```python
def korobka(link):
    page = requests.get("https://kolesa.kz"+link)
    content = page.content.decode("utf-8").lower()
    soup = BeautifulSoup(page.content, "html.parser")
    r = "неуказан"
    if "автомат" in content:
        r = "автомат"
    if "типтроник" in content:
        r = "типтроник"
    return r
```

```python
def dvigatel(link):
    page = requests.get("https://kolesa.kz"+link)
    content = page.content.decode("utf-8").lower()
    soup = BeautifulSoup(page.content, "html.parser")
    r = "не указан"
    if "3" in content:
        r = "3"
    if "2" in content:
        r = "2"
    if "1.5" in content:
        r = "1.5"
    if "2.9" in content:
        r = "2.9"
    if "2.97" in content:
        r = "2.97"
        
```

```python
def privod(link):
    page = requests.get("https://kolesa.kz"+link)
    content = page.content.decode("utf-8").lower()
    soup = BeautifulSoup(page.content, "html.parser")
    r = "не указан"
    if "задний" in content:
        r = "задний"
    if "полный" in content:
        r = "полный"

    
```

```python
prices = []
korobkas = []
dvigatels = []
privods = []
for link in all_links:
    print(link)
    prices.append(the_price(link))
    korobkas.append(korobka(link))
    dvigatels.append(dvigatel(link))
    privods.append(privod(link))
```

```python
df = pd.DataFrame({
    "car": all_cars,
    "korobka": korobkas,
    "dvigatel": dvigatels,
    "privod": privods,
    "price": price,
})
```
